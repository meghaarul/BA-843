## 4. Data

This dataset provides historical information on stocks traded on U.S. financial markets from 2013 to 2023. It covers the OHLCV data of stocks and contains information for daily, weekly, and monthly. The dataset contains 514 stocks and 1,298 columns, totaling 25 GB of data, and provides broad insights into the stock market dynamics of the U.S. stock market over years. Accessing this dataset can offer valuable insights into stock performance, trends, trading volumes, market capitalization, price movements, and much more.

As for deeper analysis, we join other information from the Global Industry Classification Standard (GICS) code, which consists of 11 sectors, 25 industry groups, 74 industries, and 163 sub-industries for each stock. This additional classification allows us to perform more granular analysis and understand the performance of stocks within specific industry sectors.

The Standard Industrial Classification was a system for classifying industries by a four-digit code as a method of standardizing industry classification for statistical purposes across agencies.
The Global Industry Classification Standard (GICS) is a method for assigning companies to a specific economic sector and industry group that best defines its business operations. GICS is used by investors and analysts to identify, compare, and contrast a firm's competitors.

## 6. Load the data

```python
# Set the path for the file
data_1 = "gs://us_stocks_data/D1/"
data_2 = "gs://us_stocks_data/Ticker/"
```

```python
# Read the data: ticker, category, stock
df_ticker = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(data_2 + "Ticker.csv")

df_category = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(data_2 + "category.csv")

df_stock = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(data_1 + "*.csv")
```

## 7. Data preparing and Data cleaning

**Stock data**

```python
# Add new column to df_stock ("stock_name")
# Add new columns to df_stock (change the format of "datetime" and add "year", "month", "date")

df_stock = df_stock.withColumn("stock_name", regexp_extract(input_file_name(), r"([^/]+)\.US_D1\.csv$", 1))
df_stock = df_stock.withColumn("datetime", to_date("datetime", "yyyy-MM-dd HH:mm:ss")) \
               .withColumn("year", year("datetime")) \
               .withColumn("month", month("datetime")) \
               .withColumn("day", dayofmonth("datetime"))

print("The total rows in the stock data is", df_stock.count())
```

Given the extensive size of the Stock data comprising 2,762,882 rows and 1,292 columns, conducting the initial analysis and merging it with additional datasets for further column expansion would demand significant computational resources and time. Hence, we have chosen to focus on selecting the first seven columns for this step, as they encapsulate the most vital information for our analysis. For the subsequent steps, we will continue exploring the remaining data to conduct deeper exploration and analysis.

```python
# Select subset of the stock data to do analysis
df_new = df_stock.select('datetime', 'open', 'high', 'low', 'close', 'volume', 'stock_name', "year", "month", "day")

# Create a new column "year_month" (for plotting)
df_new = df_new.withColumn('Year_Month',F.date_format("datetime", "yyyy-MM"))
```

**Ticker data**

```python
# Drop duplicates
df_ticker = df_ticker.dropDuplicates()
print("The total rows in the ticker data is", df_ticker.count())
```

**Category data**

```python
# Select columns from category table
df_category = df_category[['Sector', 'Sector_name','Industry_Group','Industry_Group_name','Industry','Industry_name']]

# Filter out the two outlier Industry value
df_category = df_category.filter(~df_category.Industry.isin(['  leasing  institutions','None']))

# Drop duplicates
df_category = df_category.dropDuplicates()

print("The total rows in the category data is", df_category.count())
```

**Merge the data**

```python
# First, merge the df_ticker and df_category table (join the two tables base on GICS_3 == Industry)

df_ticker = df_ticker.join(df_category, on = [df_ticker['GICS_3'] == df_category['Industry']], how = 'left')
```

Due to the issue that some of the stocks in the stock data do not have related ticker and categorical data, we decided to use "inner join" instead in order to ensure comprehensive data integration.

```python
# Second, merge the df_ticker and df_stock table (join the two tables base on GICS_3 == Industry)

df_data = df_new.join(df_ticker, on = [df_new['stock_name'] == df_ticker['Ticker']], how = 'inner')
df_data = df_data.withColumn('Year_Month',F.date_format("datetime", "yyyy-MM"))
```

#### **Q1. How has the average closing price of stocks in different sectors changed over time from 1998 to 2023?**

```python
# Group data by year-month and section name
df_group = df_data.groupBy('Year_Month','Sector_name').agg(F.avg('close').alias('Average Close Price')).orderBy(F.asc('Year_Month')).toPandas()
df_group['Year_Month'] = pd.to_datetime(df_group['Year_Month'],format='%Y-%m')

# Pivot the DataFrame to have Year_Month as index, columns as Sector_names and values as Average Open Price
pivoted_df = df_group.pivot(index='Year_Month', columns='Sector_name', values='Average Close Price')
```

```python
# Create a colormap based on the number of sector names
colormap = mpl.colormaps['Paired']
colors = colormap(np.linspace(0, 1, len(pivoted_df.columns)))

# Draw a line plot for the close price of different sections over time
plt.figure(figsize=(15, 8))
for i, column in enumerate(pivoted_df.columns):
    plt.plot(pivoted_df.index, pivoted_df[column], label=column, color=colors[i])

# Convert the index to datatime format
years = pivoted_df.index.year.unique()
# Set the x ticks as each year instead of year & month
plt.xticks(ticks=[pd.to_datetime(f'{year}-01-01') for year in years], labels=years)

plt.xlabel('Year')
plt.ylabel('Average Close Price')
plt.title('Average Close Price of Each Sector Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

#### **Q2. Which company has had the highest average volume for each of the years from 1998 to 2023?**

```python
#group by sector name, stock name and year and then finding the average volume
avg_volume_df = df_data.groupby('year', 'Sector_name', 'stock_name').agg(F.avg('volume').alias('avg_volume'))

#using partition by to rank by year
rank_window = Window.partitionBy('year').orderBy(F.desc('avg_volume'))
ranked_df = avg_volume_df.withColumn('rank', F.rank().over(rank_window))

#filtering the top stock for each of the years
top_stocks_df = ranked_df.filter(F.col('rank') == 1)
top_stocks_df.show(3)
```

```python
# Adding a new column 'year' to the DataFrame
df_data = df_data.withColumn('year', F.year('datetime'))
# Grouping the DataFrame df_data by 'year' and 'sector_name'and calculating the average of column high for each group
avg_high_per_year_sector = df_data.groupBy('year', 'sector_name') \
    .agg(F.avg('high').alias('avg_high'))
# Setting a window function to partition data by 'year' and order by average high per year sector in descending order
window = Window.partitionBy('year').orderBy(avg_high_per_year_sector['avg_high'].desc())
# Adding a new column 'rank' to the DataFrame avg_high_per_year_sector using window function rank
ranked_avg_high_per_year_sector = avg_high_per_year_sector.withColumn('rank', F.rank().over(window))
# Filtering the DataFrame to select only rows where rank is equal to 1, selecting relevant columns, and ordering by 'year'

result = ranked_avg_high_per_year_sector.filter(ranked_avg_high_per_year_sector['rank'] == 1) \
    .select('year', 'sector_name', 'avg_high') \
    .orderBy('year')
# Display the rows
result.show(3)
```

#### **Q5. Predicting MFST stock price using GBTRegressor and RFRegressor**

```python
# Stock price Regression
df_regression = df_it.select('datetime','open','high','low','close','volume','stock_name','Year_Month')
df_regression.show(5)
```

```python
df_regression.groupBy('stock_name').agg(F.avg('volume').alias('average_volume')).orderBy(F.desc('average_volume')).show()
```

```python
df_msft = df_regression.filter((F.col('stock_name') == 'MSFT') & (F.col("Year_Month") >= F.lit("2019-01")) & (F.col("Year_Month") <= F.lit("2024-01")))
df_msft.limit(3).toPandas()
```

```python
from pyspark.sql.window import Window
w = Window.partitionBy(F.year("datetime"), F.month("datetime")).orderBy("datetime")

# previous day's close price
df_msft = df_msft.withColumn("lag_close", F.lag(df_msft.close).over(w))

# daily return - how much did the price change in percentage from opening to closing
df_msft = df_msft.withColumn("daily_return", (df_msft.close - df_msft.open) / df_msft.open)

# intra-day volatility - difference between the highest and lowest prices of the day
df_msft = df_msft.withColumn("intra_day_volatility", df_msft.high - df_msft.low)

# daily volatility - change in closing price from the previous day
df_msft = df_msft.withColumn("daily_volatility", df_msft.close - df_msft.lag_close)

# 7-day moving average for the closing prices
df_msft = df_msft.withColumn("7_day_avg_close", F.avg(df_msft.close).over(w.rowsBetween(-6, 0)))

# 30-day moving average for the closing prices
df_msft = df_msft.withColumn("30_day_avg_close", F.avg(df_msft.close).over(w.rowsBetween(-29, 0)))
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import GBTRegressor
from pyspark.ml.evaluation import RegressionEvaluator
# drop the na from lagging and daily_volatility
df_msft = df_msft.dropna()
# Identify feature columns
feature_columns = ["open", "high", "low", "volume", "daily_return",
                   "intra_day_volatility", "daily_volatility", "7_day_avg_close",
                   '30_day_avg_close']
# Fit into the dataframe
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
df_assembled = assembler.transform(df_msft)

# Spilt the data into 80%, 20%
train_count = int(df_assembled.count() * 0.8)
train_data = df_assembled.orderBy("datetime").limit(train_count)
test_data = df_assembled.subtract(train_data)

# Build and Train model
gbt = GBTRegressor(labelCol="close", featuresCol="features", maxIter=50)
model = gbt.fit(train_data)
gbt_prediction = model.transform(test_data)

# Model Evaluation - RMSE, MAE and R-squared (R2)
for metric in ['rmse','mae', 'r2']:
    evaluator = RegressionEvaluator(labelCol="close", predictionCol="prediction", metricName=metric)
    value = evaluator.evaluate(gbt_prediction)
    print(f"{metric.upper()}: {value}")
```

```python
# Transform gbt prediction to dataframe
gbt_prediction = gbt_prediction.orderBy('datetime')
gbt_prediction = gbt_prediction.withColumn('gbt_prediction',F.col('prediction')).select(F.col('datetime'),F.col('close'),F.col('gbt_prediction'))

gbt_preds_pd = gbt_prediction.toPandas()
gbt_preds_pd['datetime'] = pd.to_datetime(gbt_preds_pd['datetime'])
gbt_preds_pd.head()
```

```python
from pyspark.ml.regression import RandomForestRegressor

# Build and Train model
rf = RandomForestRegressor(featuresCol="features", labelCol="close")
model = rf.fit(train_data)
rf_prediction = model.transform(test_data)

# Model Evaluation - RMSE, MAE and R-squared (R2)
for metric in ['rmse','mae', 'r2']:
    evaluator = RegressionEvaluator(labelCol="close", predictionCol="prediction", metricName=metric)
    value = evaluator.evaluate(rf_prediction)
    print(f"{metric.upper()}: {value}")
```

```python
# Transform gbt prediction to dataframe
rf_prediction = rf_prediction.orderBy('datetime')
rf_prediction = rf_prediction.withColumn('rf_prediction',F.col('prediction')).select(F.col('rf_prediction'),F.col('datetime'))
rf_prediction_pd = rf_prediction.toPandas()
rf_prediction_pd['datetime'] = pd.to_datetime(rf_prediction_pd['datetime'])

preds = pd.merge(gbt_preds_pd,rf_prediction_pd, on = 'datetime')
preds.head()
```

```python
start_date = pd.to_datetime('2022-01-01')
end_date = pd.to_datetime('2023-12-31')
mask = (preds['datetime'] >= start_date) & (preds['datetime'] <= end_date)
preds = preds.loc[mask]

# Plot the actual vs predicted values for closing prices
plt.figure(figsize=(20, 6))
plt.plot(preds["datetime"], preds["close"], label='Actual', color='blue')
plt.plot(preds["datetime"], preds["gbt_prediction"], label='gbt_prediction', color='red', alpha=0.5)
plt.plot(preds["datetime"], preds["rf_prediction"], label='rf_prediction', color='green', alpha=0.5)

plt.title('Actual vs Predicted Closing Prices')
plt.xlabel('Days')
plt.ylabel('Closing Price')
plt.legend()
plt.show()
```

#### **Analysis:**
The graph presents a comparison between actual stock closing prices of MFST and predictions made by two different machine learning models: Gradient Boosting Trees (GBT) and Random Forest Regressor (RF). The blue line represents the actual stock prices, the red line corresponds to the predictions from the GBT model, while the green line represents the RF model's predictions.

Both models appear to track the actual prices closely, following the same general trends and capturing the overall direction of stock price movements. However, there are sections of the graph where the models diverge from the actual prices, especially at the later periods of time, which represents the two model both demonstrated overfitting.

The GBT predictions closely mirror the actual price with a bit more fluctuation, suggesting a possibly higher variance. The RF predictions, while also close to the actual price, seem to be smoother, indicating that the model may be more conservative and less sensitive to short-term fluctuations.

Overall, GBT model has a lower RMSE of 15.05 and higher R2 score of 0.86, compared with RF's RMSE of 15.52 and R2 score of 0.85. So we can indicate that GBT model tends to perform better than RF in this case.

#### **Q3. Predicting WMT stock price using PCA and GBTRegressor**

```python
df_walmart = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(data_1 + "WMT.US_D1.csv")
```

```python
df_walmart = df_walmart.withColumn("datetime", to_date("datetime", "yyyy-MM-dd HH:mm:ss")) \
               .withColumn("year", year("datetime")) \
               .withColumn("month", month("datetime")) \
               .withColumn("day", dayofmonth("datetime"))
```

```python
wmt = df_walmart.toPandas()
wmt.head(3)
```

```python
feature_columns = wmt.columns.tolist()[1:]
```

```python
#Fill null values
filled_df = df_walmart.fillna(0)
```

```python
# Create shift_price for the use of prediction.
shifted_col_name = "close_price_shifted"
window_spec = Window.orderBy("datetime")
df_with_shifted = filled_df.withColumn(shifted_col_name, lead("close").over(window_spec))
# I asked chatgpt how to shift -1
```

```python
df_with_shifted=df_with_shifted.dropna()
```

```python
#Assmble features for PCA
feature_columns = wmt.columns.tolist()[1:]
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
df_walmart3 = assembler.transform(df_with_shifted)
```

```python
#Train test split
train_ratio = 0.8
train_size = int(train_ratio * df_walmart3.count())
train_data = df_walmart3.limit(train_size)
test_data = df_walmart3.subtract(train_data)
```

```python
#Since there are more than 1200 features, we use PCA to deal with it
num_principal_components = 5
pca = PCA(k=num_principal_components, inputCol="features", outputCol="pca_features")
```

```python
#Conduct PCA
pca_model = pca.fit(train_data)
pca_train_data = pca_model.transform(train_data)
pca_test_data = pca_model.transform(test_data)
```

```python
#Use GBTRegressor as model to predict the close price of Walmart.
gbt = GBTRegressor(featuresCol="pca_features",labelCol='close_price_shifted', maxIter=10)
model = gbt.fit(pca_train_data)
```

```python
predictions_train = model.transform(pca_train_data)
```

```python
#Calculate RMSE for train to check it its overfitting
evaluator = RegressionEvaluator(
    labelCol="close_price_shifted", predictionCol="prediction", metricName="rmse")
rmse_train = evaluator.evaluate(predictions_train)
print("Root Mean Squared Error (RMSE) on train data = %g" % rmse_train)
```

```python
#Calculate MAE for train to check it its overfitting
evaluator2 = RegressionEvaluator(
    labelCol="close_price_shifted", predictionCol="prediction", metricName="mae")
mae_train = evaluator2.evaluate(predictions_train)
print("Mean Absolute Error (MAE) on train data = %g" % mae_train)
```
